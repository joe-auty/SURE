```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
```

```python

N1,N2=400,300
var1,var2=4,1
mu1,sigma1=-2,var1**0.5
mu2,sigma2= 4,var2**0.5
x1=(np.random.normal(mu1,sigma1, size=N1))
x2=(np.random.normal(mu2,sigma2, size=N2))

```

```python
x=np.zeros(len(x1)+len(x2))
x[0:len(x1)]=x1
x[len(x1):]=x2
plt.plot(x,'ro')
```

```python
# finding z score, z=(x-mu)/sigma 
z=(x-mu1)/sigma1
plt.plot(z,'ro')
#mahalanobis distance 
#finding covariance 
np.cov(x,rowvar=0)
delta=(x-mu1)

```

```python
#mean and var for batch of data 
#data points 100 to 150
mu_err=np.mean(x1[100:110])-mu1
var_err=np.var(x1[100:110])-var1
print(mu_err)
print(var_err)
#different approach 
mu_err=np.mean(x1[0:N1:int(N1/10)])-mu1
var_err=np.var(x1[0:N1:int(N1/10)])-var1
print(mu_err)
print(var_err)
#slightly more accurate 
```

```python
#repeated but for other data set
#mean and var for batch of data 
#data points 100 to 150
mu_err=np.mean(x2[100:110])-mu2
var_err=np.var(x2[100:110])-var2
print(mu_err)
print(var_err)
#different approach 
mu_err=np.mean(x2[0:N1:int(N2/10)])-mu2
var_err=np.var(x2[0:N1:int(N2/10)])-var2
print(mu_err)
print(var_err)
#more accurate when the actual variance of the data set is lower, intuitively 
```
